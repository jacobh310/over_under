```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = {}
for year in range(2010,2017):
  data[year] = pd.read_csv(f'https://raw.githubusercontent.com/jacobh310/over_under/master/data_cleaning/clean_data/clean_data_{year}.csv')
  # data[year] = pd.read_csv(f'/content/clean_data_{year}.csv')
```

```python
df = pd.concat([data[year] for year in range(2010,2017)])
```

```python
over_counts = df['Over'].value_counts()

plt.figure(figsize=(8,5))
ax = sn.barplot(x=df['Over'].value_counts().index, y=df['Over'].value_counts())
ax.set_title('Over Under')
ax.set_xticklabels(['Under','Over'])
plt.show()
```

```python
df_runs_ou = df.groupby('Home Team')[['Total Runs','Home Close OU']].mean().reset_index()
df_runs_ou_melt = pd.melt(df_runs_ou, id_vars='Home Team')

fig, ax = plt.subplots(figsize=(15,9))
sn.barplot(x='Home Team', y='value', hue='variable', data=df_runs_ou_melt,ax=ax)
plt.show()
```

```python
df_era = df.groupby('Home Team')[['Home ERA','Visit ERA']].mean().reset_index()
df_era_melt = pd.melt(df_era, id_vars='Home Team')

fig, ax = plt.subplots(figsize=(15,9))
sn.barplot(x='Home Team', y='value', hue='variable', data=df_era_melt,ax=ax)
plt.show()
```
